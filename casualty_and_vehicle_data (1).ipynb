#**Data Preparation**

```python
# Import the dataset
df_casual=pd.read_csv("/content/drive/MyDrive/casualty.csv")
df_veh=pd.read_csv("/content/drive/MyDrive/vehicle.csv")
df=pd.read_csv("/content/drive/MyDrive/accident.csv")
```

```python
df_casual.head()
```

```python
casualty_df_2020 = df_casual[df_casual['accident_year'] == 2020]
casualty_df_2020.columns
```

```python
vehicle_df_2020 = df_veh[df_veh['accident_year'] == 2020]
vehicle_df_2020.columns
```

```python
accident_df_2020 = df[df['accident_year'] == 2020]
accident_df_2020.head()
```

```python
# Make a copy of the DataFrame to avoid SettingWithCopyWarning
accident_2020_df = accident_df_2020.copy()

# Convert 'date' column to datetime with the correct format
accident_2020_df['date'] = pd.to_datetime(accident_2020_df['date'], format='%d/%m/%Y')

# Extract day, month, and year into new columns
accident_2020_df['Day'] = accident_2020_df['date'].dt.day
accident_2020_df['Month'] = accident_2020_df['date'].dt.month
accident_2020_df['Year'] = accident_2020_df['date'].dt.year
accident_2020_df["time"] = accident_2020_df["time"].astype("datetime64[ns]")
accident_2020_df.head()
```

```python
accident_2020_df.info()
```

```python
# List of columns to fill "-1" with the mode
columns_to_fill = [
    'speed_limit', 'junction_detail', 'junction_control', 'second_road_class', 'second_road_number',
    'pedestrian_crossing_human_control', 'pedestrian_crossing_physical_facilities', 'light_conditions',
    'weather_conditions', 'road_surface_conditions', 'special_conditions_at_site', 'carriageway_hazards',
    'urban_or_rural_area', 'did_police_officer_attend_scene_of_accident', 'trunk_road_flag','local_authority_district'
]

# Iterate through each column and fill "-1" values with the mode
for column in columns_to_fill:
    mode_value = accident_2020_df[column].mode()[0]
    accident_2020_df[column] = accident_2020_df[column].replace(-1, mode_value)

# Check the changes
accident_2020_df.head()
```

```python
# Select the required columns from vehicle_df
vehicle_columns_needed = [
    'accident_year', 'vehicle_manoeuvre', 'vehicle_direction_from', 'vehicle_direction_to',
    'skidding_and_overturning', 'first_point_of_impact', 'vehicle_type', 'vehicle_reference',
    'vehicle_left_hand_drive', 'driver_imd_decile', 'driver_home_area_type', 'age_of_driver',
    'sex_of_driver', 'engine_capacity_cc', 'age_of_vehicle', 'journey_purpose_of_driver',
    'towing_and_articulation'
]
vehicle_2020_df = vehicle_df_2020[vehicle_columns_needed]
```

```python
vehicle_2020_df["vehicle_reference"].unique()
```

```python
# List of columns to fill "-1" with the mode
columns_to_fill = [
       'vehicle_manoeuvre', 'vehicle_direction_from',
       'vehicle_direction_to', 'skidding_and_overturning',
       'first_point_of_impact',
       'vehicle_left_hand_drive',  'driver_home_area_type',
       'sex_of_driver',
        'journey_purpose_of_driver',
       'towing_and_articulation']


# Iterate through each column and fill "-1" values with the mode
for column in columns_to_fill:
    mode_value = vehicle_2020_df[column].mode()[0]
    vehicle_2020_df[column] = vehicle_2020_df[column].replace(-1, mode_value)
```

```python
# List of columns to fill "-1" with the mean
columns_to_fill_mean = [
    'age_of_vehicle',
    'engine_capacity_cc',
    'age_of_driver',
    'driver_imd_decile'
]

# Iterate through each column and fill "-1" values with the mean
for column in columns_to_fill_mean:
    mean_value = vehicle_2020_df[column].replace(-1, pd.NA).mean()
    vehicle_2020_df[column] = vehicle_2020_df[column].replace(-1, mean_value)
```

```python
# Select the required columns from casualty_df
casualty_columns_needed = [
    'accident_year', 'sex_of_casualty', 'age_of_casualty', 'age_band_of_casualty', 'casualty_severity',
    'car_passenger', 'bus_or_coach_passenger', 'pedestrian_location', 'pedestrian_movement',
    'pedestrian_road_maintenance_worker', 'casualty_home_area_type', 'casualty_imd_decile'
]
casualty_2020_df = casualty_df_2020[casualty_columns_needed]
```

```python
# List of columns to fill "-1" with the mode
columns_to_fill = [
       'accident_year', 'sex_of_casualty',
       'car_passenger',
       'bus_or_coach_passenger', 'pedestrian_location', 'pedestrian_movement',
       'pedestrian_road_maintenance_worker', 'casualty_home_area_type',
       'casualty_imd_decile']


# Iterate through each column and fill "-1" values with the mode
for column in columns_to_fill:
    mode_value = casualty_2020_df[column].mode()[0]
    casualty_2020_df[column] = casualty_2020_df[column].replace(-1, mode_value)
```

```python
# Replace -1 in age_band_of_casualty and age_of_casualty with NaN to make it easier to compute the mean
casualty_2020_df[['age_band_of_casualty', 'age_of_casualty']] = casualty_2020_df[['age_band_of_casualty', 'age_of_casualty']].replace(-1, pd.NA)

# Compute the mean of each column
mean_age_band = casualty_2020_df['age_band_of_casualty'].mean()
mean_age_of_casualty = casualty_2020_df['age_of_casualty'].mean()

# Fill NaN values with the computed mean
casualty_2020_df['age_band_of_casualty'] = casualty_2020_df['age_band_of_casualty'].fillna(mean_age_band)
casualty_2020_df['age_of_casualty'] = casualty_2020_df['age_of_casualty'].fillna(mean_age_of_casualty)
```

```python
import numpy as np
# code to rempove 37.31211374 from the age_of_casualty column
# Replace the specific value with NaN
casualty_2020_df['age_of_casualty'].replace(37.31211374, np.nan, inplace=True)

# Drop the rows with NaN values
casualty_2020_df.dropna(subset=['age_of_casualty'], inplace=True)
```
